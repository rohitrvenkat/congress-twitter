```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in twitter accounts
twitter_df = pd.read_csv("../data/twitter-accounts-summary.csv")
# twitter_df.head()
```

```python
# Plot highest follower counts
data = twitter_df.sort_values(by = 'followers', ascending = False).head(25)

plt.figure(figsize = (6, 6))
sns.set_theme(style = 'whitegrid')
sns.barplot(data = data, x = 'followers', y = 'twitter_account', hue = 'party',
            palette = ["#1F78B4", "#E31A1C"], dodge = False, alpha = 0.8)
plt.xlabel('Followers')
plt.ylabel('')
plt.title("Most Followed Twitter Accounts", fontsize = 12)
plt.legend([],[], frameon = False)
plt.show()
```

```python
# Plot distribution of follower counts
sns.set_theme(style = 'ticks')
sns.histplot(data = twitter_df, x = 'followers', 
             log_scale = True, element = 'step',
             hue = 'party', hue_order = ['D', 'R'], 
             palette = ["#1F78B4", "#E31A1C"], alpha = 0.6);
plt.xlabel('Followers')
plt.ylabel('Twitter Accounts')
plt.legend([],[], frameon = False)
plt.show()
```
